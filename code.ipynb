```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("R_Module_Day_12.1_Network_Intrusion_Train_data.csv",delimiter=',')
```

```python
data.head()
```

```python
class1=data["class"]
#data.head()
data.drop("class",axis=1,inplace=True)
```

```python
def change1(s):
    if(s=="tcp"):
        return 0
    elif(s=="udp"):
        return 1
    else:
        return 2
```

```python
def change2(s):
    if(s=="normal"):
        return 1
    else:
        return 0
```

```python
data["protocol_type_"]=data.protocol_type.apply(change1)
```

```python
data["protocol_type"]=data["protocol_type_"]
```

```python
data.drop("service",axis=1,inplace=True)
```

```python
data.drop("flag",axis=1,inplace=True)
```

```python
output=class1.apply(change2)
```

```python
output.shape
```

```python
test=pd.read_csv('R_Module_Day_12.2_Network_Intrusion_Test_data.csv',delimiter=',')
```

```python
test.drop("service",axis=1,inplace=True)
test.drop("flag",axis=1,inplace=True)
```

```python
test["protocol_type_"]=test.protocol_type.apply(change1)
test["protocol_type"]=test["protocol_type_"]
```

```python
test.head()
```

```python
from sklearn import svm
clf1=svm.SVC()
```

```python
clf1.fit(data,output)
```

```python
predict1=clf1.predict(test)
```

```python
predict1=list(map(str, predict1))
```

```python
predict_final=[]
for i in range(len(predict1)):
    if(predict1[i]=='1'):
        predict_final.append("normal")
    else:
        predict_final.append("anomaly")
```

```python
predict_final
```

```python
np.savetxt("prediction_R_Module.csv",predict_final,fmt="%s")
```
